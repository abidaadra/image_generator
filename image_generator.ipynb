```python
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32, use_auth_token=True)
```

```python
prompt=("A cat playing soccer,""full HD 4K resolution")
output = pipe(prompt, guidance_scale=7.5)
print(output.keys())
```

```python
# if output contains multiple images
for image_data in output['images']:
    plt.axis("off")
    plt.imshow(image_data)
    plt.show()
# if output contains single image
#image_data = output['images'][0]
#plt.axis("off")
#plt.imshow(image_data)
#plt.show()

```
